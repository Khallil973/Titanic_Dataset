```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a Pandas DataFrame
titanic_df = pd.read_csv('/content/titanic/train.csv')

# Visualize the distribution of 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Add mean and median lines
age_mean = titanic_df['Age'].mean()
age_median = titanic_df['Age'].median()
plt.axvline(age_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {age_mean:.2f}')
plt.axvline(age_median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {age_median}')

# Show legend
plt.legend()

# Save the plot to a file
plt.savefig('/content/titanic_age_distribution.png')

# Display the plot
plt.show()
```

```python
# Descriptive statistics
summary_stats = titanic_df['Age'].describe()

# Create a text report
report = f"""
**Age Distribution Analysis Report**

- Mean Age: {age_mean:.2f}
- Median Age: {age_median:.2f}

Descriptive Statistics:
{summary_stats}
"""

# Print the report
print(report)

# Save the report to a text file
with open('/content/titanic_age_distribution_report.txt', 'w') as file:
    file.write(report)
```
